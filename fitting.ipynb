# Curve Fitting

```python
import numpy as np
import scipy.stats as st
import sklearn.linear_model as lm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = lambda x: x**3 - x**2 + np.exp(3 * x)
```

```python
x_tr = np.linspace(-1., 1, 200)
y_tr = f(x_tr)
```

```python
x = np.array([-1, -.9, -.8, -.5, -.2, -.1, 0, .1, .2, .5, .8, .9, 1])
y = f(x) + np.random.randn(len(x))
```

```python
plt.plot(x_tr, y_tr, '--g')
plt.plot(x, y, 'ob', marker='$○$')
plt.xlim(-1, 1)
```

```python
lr = lm.LinearRegression()
lr.fit(x[:, np.newaxis], y)
y_lr = lr.predict(x_tr[:, np.newaxis])
```

```python
plt.plot(x_tr, y_lr, 'r');
plt.plot(x_tr, y_tr, '--g')
plt.plot(x, y, 'ob', marker='$○$')
plt.xlim(-1, 1);
plt.ylim(y.min()-1, y.max()+1);
plt.title("Linear regression");
```

```python
lrp = lm.LinearRegression()

deg = 4
lrp.fit(np.vander(x, deg + 1), y);
y_lrp = lrp.predict(np.vander(x_tr, deg + 1))
plt.plot(x_tr, y_lrp, 'r', label='degree ' + str(deg));
plt.legend(loc=2);
plt.xlim(-1., 1.);
plt.ylim(y.min()-1, y.max()+1);
print(' '.join(['%.2f' % c for c in lrp.coef_]))
plt.plot(x_tr, y_tr, '--g')
plt.plot(x, y, 'ob', marker='$○$')
plt.title("Linear regression");
```

```python
lrp = lm.LinearRegression()

deg = 15
lrp.fit(np.vander(x, deg + 1), y);
y_lrp = lrp.predict(np.vander(x_tr, deg + 1))
plt.plot(x_tr, y_lrp, 'r', label='degree ' + str(deg));
plt.legend(loc=2);
plt.xlim(-1., 1.);
plt.ylim(y.min()-1, y.max()+1);
print(' '.join(['%.2f' % c for c in lrp.coef_]))
plt.plot(x_tr, y_tr, '--g')
plt.plot(x, y, 'ob', marker='$○$')
plt.title("Linear regression");
```
